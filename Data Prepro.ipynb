```python
import pandas as pd
import numpy as np
import math
```

```python
file_path = 'Fosse_Chichawa_EC_Olivier_30min_2019-.xlsx'
# Charger le fichier Excel avec l'extension .xlnx
df = pd.read_excel(file_path, engine='openpyxl')
# Afficher les premières lignes du DataFrame
df
```

```python
# Sélectionner les colonnes à conserver
colonnes_gardees = ['Date', 'Heure','RH','AirTC','G','Rn_Avg','Le_flux','Ux','Uy','Uz','short_up_Avg']

# Utiliser loc pour sélectionner les colonnes
Our_data = df.loc[:, colonnes_gardees]

# Afficher le dataframe résultant
Our_data
```

```python
Our_data.info()
```

```python
# Compter le nombre de valeurs nulles dans chaque colonne
null_counts = Our_data.isnull().sum()
null_counts
```

```python
# Liste des colonnes où remplir les valeurs null
columns_to_fill = ['RH', 'AirTC', 'G', 'Rn_Avg', 'Le_flux', 'Ux', 'Uy', 'Uz','short_up_Avg']
# Convertir les colonnes numériques en type numérique
Our_data['Ux'] = pd.to_numeric(Our_data['Ux'], errors='coerce')
Our_data['Uy'] = pd.to_numeric(Our_data['Uy'], errors='coerce')
Our_data['Uz'] = pd.to_numeric(Our_data['Uz'], errors='coerce')
Our_data['Le_flux'] = pd.to_numeric(Our_data['Le_flux'], errors='coerce')
```

```python
null_counts = Our_data.isnull().sum()
null_counts
```

```python
# Define the function to calculate the square root of the sum of squares
def calculate_ws(row):
    return np.sqrt(row['Ux']**2 + row['Uy']**2 + row['Uz']**2)

# Apply the function to each row and store the result in a new column 'Ws'
Our_data['Ws'] = Our_data.apply(calculate_ws, axis=1)

# Print the updated dataframe
Our_data
```

```python
# Définir la fonction pour calculer Psat
def calculatePsat(row):
    return 0.6112 * math.exp((17.62 * row['AirTC']) / (row['AirTC'] + 243.12))

# Appliquer la fonction à chaque ligne et stocker le résultat dans une nouvelle colonne 'Psat'
Our_data['Psat'] = Our_data.apply(calculatePsat ,axis=1)
Our_data
```

```python
# Définir la fonction pour calculer Psat
def calculateP_air(row):
    return row['Psat']* (row['RH']/100)

# Appliquer la fonction à chaque ligne et stocker le résultat dans une nouvelle colonne 'Psat'
Our_data['P_air'] = Our_data.apply(calculateP_air ,axis=1)
Our_data
```

```python
# Définir la fonction pour calculer Psat
def calculateVPD(row):
    return   row['Psat']* (1 - row['RH']/100) 

# Appliquer la fonction à chaque ligne et stocker le résultat dans une nouvelle colonne 'Psat'
Our_data['VPD'] = Our_data.apply(calculateVPD ,axis=1)
Our_data
```

```python
print("Data type of 'short_up_Avg':", Our_data['short_up_Avg'].dtypes)
Our_data['short_up_Avg'] = Our_data['short_up_Avg'].astype(float)
```

```python
# Grouper les données par jour et calculer la somme de "short_up_Avg"
daily_sum = Our_data.groupby(Our_data['Date'].dt.date)['short_up_Avg'].sum()

# Diviser les valeurs par 48 pour obtenir la moyenne par tranche de 48 unités de temps
Rs = daily_sum / 48
Rs
```

```python
# Créer une colonne 'Rs' dans le DataFrame Our_data et y stocker les valeurs de Rs
Our_data['Rs'] = Our_data['Date'].dt.date.map(Rs)

# Afficher le DataFrame avec la colonne 'Rs' mise à jour
print(Our_data)
```

```python
# Sélectionner les colonnes à conserver
colonnes_gardees = ['Date', 'Heure','RH','AirTC','G','Rn_Avg','Le_flux','Ws','Psat','P_air','VPD','Rs']

# Utiliser loc pour sélectionner les colonnes
Final_Data = Our_data.loc[:, colonnes_gardees]

# Afficher le dataframe résultant
Final_Data
```
